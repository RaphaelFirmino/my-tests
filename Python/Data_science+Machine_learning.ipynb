```python
# inteligencia artificial é uma previsão do que já acontece ou que pode acontecer com base em uma ciência de dados.
# correlação (-1 a 1) é quando itens tem um comportamento praticamente iguais, exemplo quando um aumenta ou abaixa o outro item tem o msm comportamento. 
import pandas as pd
```

```python
tabela = pd.read_csv(r'/content/drive/MyDrive/Colab Notebooks/Hashtag intensivao/Aula 3/advertising.csv')
display(tabela)
tabela.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# verificar quem tem a maior correlação com vendas
# caracteristticas são os parametros utilizados para determinado resultado
# para inteligencia artificial não pode haver uma correlação alta entre as caracteristicas, como a partir de 0.8
sns.heatmap(tabela.corr(), annot=True, cmap='Wistia') # criar um gráfico e solicitar uma correlação
plt.show()

sns.pairplot(tabela)
plt.show()
```

```python
# coletar dados para uma inteligencia artificial
from sklearn.model_selection import train_test_split
```

```python
y = tabela['Vendas']
x = tabela.drop('Vendas', axis=1)

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1) # formato padrão de variaveis para essa função (as 4 variaveis)
```

```python
# metodos de inteligencia artificial: ter uma previsão de um número ou categorização(separar como A, B e etc)
# para nosso caso será utilizado a previsão, testando com: regressão linear e arvore de decisão(random forest)
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
# criar as inteligencias artificiais
modelo_regressaolinear = LinearRegression()
modelo_arvoredecisao = RandomForestRegressor()

# treinar as IAs
modelo_regressaolinear.fit(x_treino, y_treino)
modelo_arvoredecisao.fit(x_treino, y_treino)
```

```python
# avaliar os resultados e escolher qual é a melhor IA, utilizando o método R²
# R² = é a porcentagem que nossa IA consegue exlicar o que acontece na realidade
from sklearn import metrics
```

```python
# visualização gráfica das previsões
tabela_auxiliar = pd.DataFrame() # criando a tabela
tabela_auxiliar['Valor real'] = y_teste
tabela_auxiliar['Regressão linear'] = previsao_regressaolinear
tabela_auxiliar['Arvore de decisões'] = previsao_arvoredecisao

plt.figure(figsize=(15,5)) # ajustando o tamanho do gráfico
sns.lineplot(data=tabela_auxiliar)
plt.show()
```

```python
# qual investimento foi mais correlacionado com as vendas
sns.barplot(x=x_treino.columns, y=modelo_arvoredecisao.feature_importances_)
plt.show()
```

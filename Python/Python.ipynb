```python
print("\nExercise 3\n")
#from math import sqrt
a = int(input("Write the value of A: "))
b = int(input("Write the value of B: "))
c = int(input("Write the value of C: "))
 
delta = b**2 - 4*a*c
if delta > 0:

  raiz_delta = (delta**(1/2))
  x1 = (-b + raiz_delta)/2*a
  x2 = (-b - raiz_delta)/2*a
  print("The roots are", x1, "and", x2)

else:
  print("Negative delta")
```

```python
print("\nExercise 5\n")
n1 = float(input("Write the first number: "))
sinal = input("Write the signal (+,-,* or /): ")
n2 = float(input("Write the second number: "))
if sinal == "+":
    op = n1 + n2
elif sinal == "-":
    op = n1 - n2
elif sinal == "*":
    op = n1 * n2
elif sinal == "/":
    op = n1 / n2
else:
    print("Invalid signal")
 
print(op)
```

```python
# map
def dobro(x):
  return x*2
valor = [1,2,3,4,5]
valor_dobrado = map(dobro,valor)
valor_dobrado = list(valor_dobrado)
print(valor_dobrado)
```

```python
# reduce
from functools import reduce
def soma(x,y):
  return x+y
lista = [2,3,5,10,20]
soma = reduce(soma,lista)
print(soma)
```

```python
# zip
list1 = [1,2,3,4,5]
list2 = ["avocado","banana","star fruit","damascus","passion fruit"] # carambola e maracuja
list3 = ["R$ 1,00","R$ 2,00","R$ 3,00","R$ 4,00","R$ 5,00"]
for item,nome,valor in zip(list1,list2,list3):
  print(item,nome,valor)
```

```python
# arquivos python e base de dados precisam estar na msm pasta
import pandas as pd
import plotly.express as px
```

```python
# importar a base de dados para fazer a analise e tratar os dados
# Churn = cliente que cancelou algum serviço
# NaN = valor vazio
tabela = pd.read_csv(r'Nome do arquivo.csv') # .csv tbm serve para .txt
display(tabela)
print(tabela.info()) # exibir o formato dos dados
```

```python
# tratar os dados, colunas sempre deve ser tratadas primeiro e depois as linhas
# sempre analisar a porcentagem dos dados a serem deletados, se for alta é necessário corrigi-los
# coluna Unnamed é inutil, posso deletar
tabela = tabela.drop('Nome do q vai ser deletado', axis=1) # axis significa eixo e 0 para linha e 1 para coluna
display(tabela)
```

```python
# corrigir formato dos dados
tabela['Coluna q será corrigida'] = pd.to_numeric(tabela['Coluna q será corrigida'], errors='coerce') # coerce = o que não for numero ficará vazio(NaN)
print(tabela.info())
```

```python
# exluir colunas com todos valores vazios
tabela = tabela.dropna(how='all', axis=1) all = todos, any = algum
# exluir linhas com algum valor vazio
tabela = tabela.dropna(how='any', axis=0)
print(tabela.info())
```

```python
# calcular os cancelamentos
tabela['Coluna de cancelamentos'].value_counts() # dará os números
tabela['Coluna de cancelamentos'].value_counts(normalize=True).map('(:.1%)'.format) # dará em porcentagem
```

```python
# mostrar os cancelamentos através de gráficos
''' em python sempre que realizado comando for em uma tabela são analizadas as colunas e
para analisar as linhas se acrescenta .index ao nome da tabela '''
for coluna in tabela:
  grafico = px.histogram(tabela, x=coluna, color='Churn')
  grafico.show()
```

Analisar os gráficos para encontrar possíveis ligações entre o número de cancelamentos e propor idéias para melhorias.

Cliente com familia maiores tendem a cancelar menos.
  Mais pessoas utilizando nossos serviços.

Número de cancelamento é muito maior no inicio da assinatura.
   A captação pode estar sendo ruim.
   Não estão incentivados a continuar com a assinatura.
  
Cliente com fibra óptica tendem a ter mais cancelamentos.
  Pode estar acontecendo algum problema com a fibra óptica.

Quanto mais serviços o cliente tem, menos ele pede cancelamento.
  Cliente fica mais comprometido com nossa plataforma.

Contratos mensais tem mais chances de cancelar.
  Cliente não está sendo fidelizado.

Clientes que pagam com boleto tendem a cancelar mais.
  Forma de pagamento que não está trazendo bons resultados.

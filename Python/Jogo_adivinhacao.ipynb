```python
import random
```

```python
def repetir(s):
  print('\nJogo de adivinhação\n\nVocê terá 10 tentativas para adivinhar o número entre 0 e 50, boa sorte!\n')
 
  tentativas = 1
  numero = random.randint(0,50)
  while tentativas <= 10:
    print('Tentativa: ',str(tentativas)+'\n')
    entrada = int(input('Digite seu número: '))
    if entrada > numero:
      print('\nVocê errou: seu número está acima.\n')
    elif entrada < numero:
      print('\nVocê errou: seu número está abaixo.\n')
    else:
      continuar = input('\nVocê ganhou, parabéns!!!\n\nQuer jogar novamente?\nDigite S ou N: \n')
      if continuar == ('s' or 'S'):
        repetir(1)
      else:
         print('\nFim de jogo, muito obrigado!')
         break
    if tentativas == 10:
      continuar = input('Você perdeu, quer jogar novamente?\nDigite S ou N: \n')
      if continuar == ('s' or 'S'):
        repetir(1)
      else:
         print('\nFim de jogo, muito obrigado!')
    tentativas += 1
 
repetir(1)
```

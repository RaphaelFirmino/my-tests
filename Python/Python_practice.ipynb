```python
print("Fuel\n")

op = "1"
count_alcohol = 0
count_gasoline = 0
count_diesel = 0

while op != "4":
  op = input("Type a code: ")

  if op == "1":
    count_alcohol += 1
  elif op == "2":
    count_gasoline += 1
  elif op == "3":
    count_diesel += 1

print("\nThankfully!")
print("Alcohol: ", count_alcohol)
print("Gasoline: ", count_gasoline)
print("Diesel: ", count_diesel)
```

```python
print("Fibonacci Vetor\n")

n = int(input("Type a value: "))

def fib(x):
  v = [0,1,1] # [2,3,5,8,13]
  for i in range(3, x+1): # (3,8)
    v.append(v [i-2] + v [i-1]) # v[5] + v[6]
  return v[x]

saida = fib(n)

print("\nFib ("+str(n)+") = ", saida) # "+str(n)+" = retornar representacao de valor no meio de uma string
```

```python
print("Lower value\n")

x = [float(i) for i in input("Enter numbers: ").split()]

menor = min(x) # min() = menor valor
indice = x.index(menor) # .index() = localiza a posicao do item

print("Lower value = ", menor)
print("Position = ", indice)
```

```python
print("Slugs race\n")

valores = [float(i) for i in input("Enter the speeds: ").split()]

maior = max(valores)
if maior < 10:
  print("\nLevel = 1")
elif 10 <= maior < 20:
  print("\nLevel = 2")
else:
  print("\nLevel = 3")
```
